# Maps

## 2020

### On & Offshore

```python
year = '2020'
data = load_from_path(results_paths[year]['onshore'],results_paths[year]['offshore'])
data_cleanup = load_from_path(results_paths['cleanup'][year])
data = pd.concat([data,data_cleanup])
data = data.drop_duplicates(subset=['country','lat','lon','sea_node'],keep='last')
lcof_map(data,figsize=(15,15))

save_path = os.path.join(final_results_path,f'{year}.csv')
if os.path.isfile(save_path):
    print('File already exists. Make sure you want to overwrite the file before proceeding.')
else:
    assert data.duplicated(subset=['country','lat','lon','sea_node']).sum() == 0
    data.to_csv(save_path,index=False)
```
